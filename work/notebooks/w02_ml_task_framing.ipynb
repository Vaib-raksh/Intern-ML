## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can only use simple conditions, such as assuming that pages in higher position tiers always receive higher CTR. In reality, CTR is influenced by multiple factors such as search volume, impressions, content age, and other page characteristics. Machine learning can learn these combined patterns from historical data, making it more flexible than manually writing many if-else rules
