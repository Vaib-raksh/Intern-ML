## Lane: Quantifying the Value of Position

I chose this lane because I want to understand how a webpage's position in search results affects its click-through rate (CTR). In Week 1, I observed that CTR decreases as pages move to lower position tiers, but I want to measure how large that change actually is. If we can estimate the difference in CTR between position tiers, it can help content teams decide where improving a page's ranking is likely to have the greatest impact.

**Cost of a wrong recommendation:** If the expected CTR improvement is overestimated, the team may spend time and money improving pages that provide little benefit. ## Quick Look

To understand whether this research question is worth exploring, I calculated the average CTR for each position tier and compared consecutive tiers. These numbers provide an initial view of how CTR changes across search result positions and whether moving between tiers may have a meaningful impact.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4))

# how much bigger is the top tier's CTR vs the next tier down?
tiers = ctr_by_pos.index.tolist()
for i in range(len(tiers) - 1):
    ratio = ctr_by_pos[tiers[i]] / ctr_by_pos[tiers[i+1]]
    print(f"{tiers[i]} vs {tiers[i+1]}: {ratio:.2f}x the CTR")
```
